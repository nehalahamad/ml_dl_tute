```python
import matplotlib.pyplot as plt
from collections import Counter
```

```python
import rarfile
import re

rf = rarfile.RarFile("documents.rar")
file_name =[re.split('_|/', f.filename)[1] for f in rf.infolist()]
```

```python
file_name_dict = Counter(file_name)
name = file_name_dict.keys()
count = file_name_dict.values()
```

```python
plt.figure(figsize=(12,7))
plt.bar(name, count, edgecolor='black')
plt.xticks(rotation=90);
```
