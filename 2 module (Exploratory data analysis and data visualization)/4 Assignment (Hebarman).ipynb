```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('datasets_474_966_haberman.csv', header=None, names=['age', 'op_year', 'axil_nodes', 'surv_status'])
data.head()
```

```python
#a  'age' of patient at time of operation (numerical)
#b  Patient's year of operation (year - 1900, numerical)
#c  Number of positive axillary nodes detected (numerical)

#d  Survival Status
#    1 = the patient survived 5 years or longer
#    2 = the patient died within 5 years

```
