```python
def draw_line(coef,intercept, mi, ma):
    # for the separating hyper plane ax+by+c=0, the weights are [a, b] and the intercept is c
    # to draw the hyper plane we are creating two points
    # 1. ((b*min-c)/a, min) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the minimum value of y
    # 2. ((b*max-c)/a, max) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the maximum value of y
    points = np.array([[((-coef[1]*mi - intercept)/coef[0]), mi], [((-coef[1]*ma - intercept)/coef[0]), ma]])
    plt.plot(points[:,0], points[:,1])
```

# What if Data is imabalanced

<pre>
1. As a part of this task you will observe how linear models work in case of data imbalanced
2. observe how hyper plane is changs according to change in your learning rate.
3. below we have created 4 random datasets which are linearly separable and having class imbalance
4. in the first dataset the ratio between positive and negative is 100 : 2, in the 2nd data its 100:20, 
in the 3rd data its 100:40 and in 4th one its 100:80
</pre>

```python
data = []
# here we are creating 2d imbalanced data points 
ratios = [(100,2), (100, 20), (100, 40), (100, 80)]
plt.figure(figsize=(20,5))
for j,i in enumerate(ratios):
    plt.subplot(1, 4, j+1)
    X_p = np.random.normal( 0, 0.05, size=(i[0], 2) )
    X_n = np.random.normal( 0.13,0.02,size=(i[1], 2) )
    y_p = np.array([1]*i[0]).reshape(-1,1)
    y_n = np.array([0]*i[1]).reshape(-1,1)
    X = np.vstack((X_p,X_n))
    y = np.vstack((y_p,y_n))
    
    data.append([X, y])
    
    plt.scatter(X_p[:,0], X_p[:,1])
    plt.scatter(X_n[:,0], X_n[:,1], color='red')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(20,20))
c = [0.001, 1, 100]
index=1
for d in data:
    for para in c:
        svc_clf = SVC(C=para, kernel='linear').fit(d[0], d[1].ravel())
        coef = svc_clf.coef_
        intercept = svc_clf.intercept_
        mi, ma = d[0][:, 1].min(), d[0][:, 1].max()
        
        support_vectors = svc_clf.support_vectors_
        
        plt.subplot(len(data), len(c), index)
        sns.scatterplot(d[0][:, 0], d[0][:, 1], hue=d[1].ravel())
        plt.scatter(support_vectors[:,0], support_vectors[:,1], c='red', s=15, label='sv')
        plt.title('C={}  &  ratio of +ve and -ve data is: {}'.format(para, ratios[data.index(d)]))
        plt.legend()
        draw_line(coef[0], intercept, mi, ma)
        index+=1
```

## Observation:
We know that, when C decreases, bias increase and variance decrease, and when C increases, bias decrease and variance increase<br>
**Case 1: When data is in ratio of (100, 2):** since, there are very less -ve data, as the C increases biasness decreases and hyper plane is converging very slowly.<br>
**Case 2: When data is in ratio of (100, 20):** here -ve data is increases but still -ve data is very less as compared to +ve data, as the C increases biasness decreases and hyper plane is converging very slowly.<br>
**Case 3: When data is in ratio of (100, 40):** again -ve data is incresed, but still -ve data is in minority class, as the C increases biasness decreases and hyper plane is converging slowly.<br>
**Case 4: When data is in ratio of (100, 80):** here ratio of +ve and -ve data are same, as the C increases biasness decreases and hyper plane is converging very fast.
