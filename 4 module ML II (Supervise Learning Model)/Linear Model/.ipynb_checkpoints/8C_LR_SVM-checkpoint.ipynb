```python
# you should not import any other packages
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from sklearn.linear_model import SGDRegressor
```

```python
import scipy as sp
import scipy.optimize

def angles_in_ellipse(num,a,b):
    assert(num > 0)
    assert(a < b)
    angles = 2 * np.pi * np.arange(num) / num
    if a != b:
        e = (1.0 - a ** 2.0 / b ** 2.0) ** 0.5
        tot_size = sp.special.ellipeinc(2.0 * np.pi, e)
        arc_size = tot_size / num
        arcs = np.arange(num) * arc_size
        res = sp.optimize.root(lambda x: (sp.special.ellipeinc(x, e) - arcs), angles)
        angles = res.x 
    return angles
```

```python
a = 2
b = 9
n = 50

phi = angles_in_ellipse(n, a, b)
e = (1.0 - a ** 2.0 / b ** 2.0) ** 0.5
arcs = sp.special.ellipeinc(phi, e)

fig = plt.figure()
ax = fig.gca()
ax.axes.set_aspect('equal')
ax.scatter(b * np.sin(phi), a * np.cos(phi))
plt.show()
```

```python
X = b * np.sin(phi)
Y = a * np.cos(phi)

X.shape
```

```python
alphas = [0.0001, 1, 100, 1000]
outliers = [(0,2),(21, 13), (-23, -15), (22,14), (23, 14)]
plt.figure(figsize=(20,15))
index=1
for alpha in alphas:
    X_new = X
    Y_new = Y
    for outlier in outliers:
        plt.subplot(len(alphas), len(outliers), index)
        X_new = np.append(X_new, outlier[0])
        Y_new = np.append(Y_new, outlier[1])
        
        reg = SGDRegressor(alpha=alpha, eta0=0.001, learning_rate='constant',random_state=0).fit(X_new.reshape(-1, 1), Y_new.reshape(-1, 1))
        
        plt.scatter(X_new, Y_new)
        
        x1, x2 = X_new.min(), X_new.max()
        f = lambda x : reg.coef_[0] * x + reg.intercept_[0] # y=m*x+c
        
        plt.plot([x1, x2], [f(x1), f(x2)], c='red')
        plt.xlabel('alpha = {}'.format(alpha))
        index+=1
    print()
```

# Observation
**(1) -** when alpha=0.001, the regresion line is very much affected by outlier, as we can see from the above fig that, regression line is fitting the outlier very fast.

**(2) -** when alpha=1, the regresion line is slightly affected by outlier.

**(3) -** when alpha=100, the regresion line very less affected by outlier, as we can see from the above fig that, regression line is fitting the outlier very slowly.

**(4) -** when alpha=1000, the regresion line, almost have no effect of outliers.
